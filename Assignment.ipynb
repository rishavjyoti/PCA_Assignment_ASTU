# Reducing ddimensionality using PCA

```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv(r'PCA_practice_dataset.csv')
df = pd.DataFrame(data)
df
```

```python
X = df.to_numpy()
X.shape
```

# Applying PCA 

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA()
X = pca.fit_transform(X)
#Sum of all the eigen values
sum = np.cumsum(pca.explained_variance_ratio_)*100
#Thresholds
thresholds = [i for i in range (90,97+1,1)]

#List of components required to retain the varience as required
comps = [np.argmax(sum>threshold)for threshold in thresholds]
for comp,threshold in zip(comps,thresholds):
    print("Components required for {}% threshold are:{}".format(threshold,comp))
```

```python
import matplotlib.pyplot as plt
```

# Scree Plot

```python
plt.plot(comps,range(90,97+1,1),'bo-',linewidth=2)
plt.title('Scree Plot')
plt.xlabel('Principal Component')
plt.ylabel('Threshold')
plt.show()
```

# Dimensionality reduction

```python
X_orig = X

for comp,var in zip(comps,thresholds):
    pca = PCA(n_components=comp)
    X_transformed = pca.fit_transform(X_orig)
    print('Dimensionality reduction to retain {}% thresholds'.format(var))
    print('After dimensionality reduction, new shape of the dataset is :',X_transformed.shape)
```
